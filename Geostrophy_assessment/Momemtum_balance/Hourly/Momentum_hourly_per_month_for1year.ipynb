# A key term zeta

```python
# (7*12 = ~90 mins one face)
# 1.5*13 = 20 hours

dij=4

#for face in range(13):
for face in F:
    for t in T: 
        
        zeta_3 = (grid.diff(ds_raw.dyC.isel(face=face) * ds_raw.SSV.isel(face=face).isel(time=slice(t,t+nfft)),'X') - grid.diff(ds_raw.dxC.isel(face=face) * ds_raw.SSU.isel(face=face).isel(time=slice(t,t+nfft)),'Y'))/ds_raw.rAz.isel(face=face) # relative vorticity  
        zeta = grid.interp(grid.interp(zeta_3,'Y'),'X').isel(i=slice(0,None,dij), j=slice(0,None,dij))   #(zeta) at i,j
        zeta = np.real(zeta).rename('zeta')
        zeta = zeta.chunk({'time': 1,'i':1080,'j':1080})
        zeta.to_dataset().to_zarr(E_dir+'zeta_rotation_f%02d_t%04d.zarr'% (face,t), mode='w')
        print('t=%d - allready processed'%(t))
        
    print('face=%d - allready processed'%(face))
```

# Term 1: time acceleration

### du/dt and dv/dt

```python
# # 15 mins per face

#for face in range(13):
for face in F:
    
    d_SSU_dt_time = ( ds_raw['SSU'].isel(face=face).shift(time=-1) - ds_raw['SSU'].isel(face=face) )/3600
    d_SSV_dt_time = ( ds_raw['SSV'].isel(face=face).shift(time=-1) - ds_raw['SSV'].isel(face=face) )/3600
    
    for t in T: 
    
        dudt = grid.interp(d_SSU_dt_time.isel(time=slice(t,t+nfft)),'X')
        dvdt = grid.interp(d_SSV_dt_time.isel(time=slice(t,t+nfft)),'Y')    
        
        du_dt =  dudt.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face, i=slice(0,None,dij), j=slice(0,None,dij)) - dvdt.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face, i=slice(0,None,dij), j=slice(0,None,dij))
        dv_dt =  dudt.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face, i=slice(0,None,dij), j=slice(0,None,dij)) + dvdt.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face, i=slice(0,None,dij), j=slice(0,None,dij))    
        #print(du_dt)
        # store
        du_dt = np.real(du_dt).rename('du_dt')    
        du_dt = du_dt.chunk({'time': 1,'i':1080,'j':1080})
        du_dt.to_dataset().to_zarr(E_dir+'du_dt_f%02d_t%04d.zarr'% (face,t), mode='w')

        dv_dt = np.real(dv_dt).rename('dv_dt')    
        dv_dt = dv_dt.chunk({'time': 1,'i':1080,'j':1080})
        dv_dt.to_dataset().to_zarr(E_dir+'dv_dt_f%02d_t%04d.zarr'% (face,t), mode='w')        
        print('t=%d - allready processed'%(t))     
        
    print('face=%d - allready processed'%(face))        
```

# Term 3: Ageostrophic momemtum

### Terms 2-3 together

```python
# ~15 mins for one face

dij=4

#for face in range(13):
for face in F:  
    for t in T:        
        
        U_ij = grid.interp(ds_raw['SSU'].isel(face=face).isel(time=slice(t,t+nfft))  ,'X')
        V_ij = grid.interp(ds_raw['SSV'].isel(face=face).isel(time=slice(t,t+nfft))  ,'Y')

        SSU_rotation = U_ij*ds_raw.CS.isel(face=face) - V_ij*ds_raw.SN.isel(face=face)
        SSV_rotation = U_ij*ds_raw.SN.isel(face=face) + V_ij*ds_raw.CS.isel(face=face)
        SSU_rotation = np.real(SSU_rotation).rename('SSU_rotation')
        SSV_rotation = np.real(SSV_rotation).rename('SSV_rotation')
        

        g = 9.81
        fv = SSV_rotation.isel(i=slice(0,None,dij), j=slice(0,None,dij))*f_ij.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) 
        fu = SSU_rotation.isel(i=slice(0,None,dij), j=slice(0,None,dij))*f_ij.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))

        Eta = ds_raw['Eta'].isel(face=face).isel(time=slice(t,t+nfft))   
        dB_dx_raw = grid.diff(g*Eta,'X')/ds_raw.dxC.isel(face=face)     
        dB_dy_raw = grid.diff(g*Eta,'Y') /ds_raw.dyC.isel(face=face)     
        # i,j
        dB_dx_raw_ij = grid.interp(dB_dx_raw,'X')
        dB_dy_raw_ij = grid.interp(dB_dy_raw,'Y')
        # rotation
        dB_dx =  dB_dx_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) - dB_dy_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))
        dB_dy =  dB_dx_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) + dB_dy_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))
        
        
        # store
        fageo_x = -fv + dB_dx
        fageo_x = np.real(fageo_x).rename('fageo_x')
        fageo_x = fageo_x.chunk({'time': 1,'i':1080,'j':1080})
        fageo_x.to_dataset().to_zarr(E_dir+'fageo_x_f%02d_t%04d.zarr'% (face,t), mode='w')
        
        fageo_y = fu + dB_dy
        fageo_y = np.real(fageo_y).rename('fageo_y')
        fageo_y = fageo_y.chunk({'time': 1,'i':1080,'j':1080})
        fageo_y.to_dataset().to_zarr(E_dir+'fageo_y_f%02d_t%04d.zarr'% (face,t), mode='w')
        
        print('t=%d - allready processed'%(t))
        
    print('face=%d - allready processed'%(face))
```

```python
# ~15 mins for one face

dij=4

#for face in range(13):
for face in F:  
    for t in T:        
        
        U_ij = grid.interp(ds_raw['SSU'].isel(face=face).isel(time=slice(t,t+nfft))  ,'X')
        V_ij = grid.interp(ds_raw['SSV'].isel(face=face).isel(time=slice(t,t+nfft))  ,'Y')

        SSU_rotation = U_ij*ds_raw.CS.isel(face=face) - V_ij*ds_raw.SN.isel(face=face)
        SSV_rotation = U_ij*ds_raw.SN.isel(face=face) + V_ij*ds_raw.CS.isel(face=face)
        SSU_rotation = np.real(SSU_rotation).rename('SSU_rotation')
        SSV_rotation = np.real(SSV_rotation).rename('SSV_rotation')
        
        ds_zeta = xr.open_zarr(E_dir+'zeta_rotation_f%02d_t%04d.zarr'% (face,t))
        zeta_v = ds_zeta['zeta']*SSV_rotation.isel(i=slice(0,None,dij), j=slice(0,None,dij))
        zeta_u = ds_zeta['zeta']*SSU_rotation.isel(i=slice(0,None,dij), j=slice(0,None,dij))
        
        KE = (SSU_rotation**2 + SSV_rotation**2)/2 # KE
        # B represents KE here    
        dKE_dx_raw = grid.diff(KE,'X')/ds_raw.dxC.isel(face=face)     # d(KE)/dx
        dKE_dy_raw = grid.diff(KE,'Y')/ds_raw.dyC.isel(face=face)     # d(KE)/dy
        # i,j
        dKE_dx_raw_ij = grid.interp(dKE_dx_raw,'X')
        dKE_dy_raw_ij = grid.interp(dKE_dy_raw,'Y')
        # rotation 
        dKE_dx =  dKE_dx_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) - dKE_dy_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))
        dKE_dy =  dKE_dx_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) + dKE_dy_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))

        
        g = 9.81
        fv = SSV_rotation.isel(i=slice(0,None,dij), j=slice(0,None,dij))*f_ij.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) 
        fu = SSU_rotation.isel(i=slice(0,None,dij), j=slice(0,None,dij))*f_ij.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))

        Eta = ds_raw['Eta'].isel(face=face).isel(time=slice(t,t+nfft))   
        dB_dx_raw = grid.diff(g*Eta,'X')/ds_raw.dxC.isel(face=face)     
        dB_dy_raw = grid.diff(g*Eta,'Y') /ds_raw.dyC.isel(face=face)     
        # i,j
        dB_dx_raw_ij = grid.interp(dB_dx_raw,'X')
        dB_dy_raw_ij = grid.interp(dB_dy_raw,'Y')
        # rotation
        dB_dx =  dB_dx_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) - dB_dy_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))
        dB_dy =  dB_dx_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.SN.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij)) + dB_dy_raw_ij.isel(i=slice(0,None,dij), j=slice(0,None,dij))*ds_raw.CS.isel(face=face,i=slice(0,None,dij), j=slice(0,None,dij))

        
        # store
        nolinear_x = -zeta_v + dKE_dx
        nolinear_x = np.real(nolinear_x).rename('nolinear_x')
        nolinear_x = nolinear_x.chunk({'time': 1,'i':1080,'j':1080})
        nolinear_x.to_dataset().to_zarr(E_dir+'nolinear_x_f%02d_t%04d.zarr'% (face,t), mode='w')
        
        nolinear_y = zeta_u + dKE_dy
        nolinear_y = np.real(nolinear_y).rename('nolinear_y')
        nolinear_y = nolinear_y.chunk({'time': 1,'i':1080,'j':1080})
        nolinear_y.to_dataset().to_zarr(E_dir+'nolinear_y_f%02d_t%04d.zarr'% (face,t), mode='w')
        
        
        # store
        fageo_x = -fv + dB_dx
        fageo_x = np.real(fageo_x).rename('fageo_x')
        fageo_x = fageo_x.chunk({'time': 1,'i':1080,'j':1080})
        fageo_x.to_dataset().to_zarr(E_dir+'fageo_x_f%02d_t%04d.zarr'% (face,t), mode='w')
        
        fageo_y = fu + dB_dy
        fageo_y = np.real(fageo_y).rename('fageo_y')
        fageo_y = fageo_y.chunk({'time': 1,'i':1080,'j':1080})
        fageo_y.to_dataset().to_zarr(E_dir+'fageo_y_f%02d_t%04d.zarr'% (face,t), mode='w')
        
        print('t=%d - allready processed'%(t))
        
    print('face=%d - allready processed'%(face))
```

```python
latb = np.arange(-60.,60.,1.)
latc = (latb[:-1]+latb[1:])*.5

def _zhist(x, lat, weights=True):
    # an alternative would be to create xarray dataarrays and use groupby_bins
    # https://github.com/pydata/xarray/issues/1765
    if weights:
        w = x
    else:
        w = np.ones_like(x)
    h, edges = np.histogram(lat, bins=latb, weights=w, density=False)
    return h[None,:]

def zaverage(v):
    dims = ['i','j']
    v = v.fillna(0.)
    lat = v.YC+0.*v
    #
    S = xr.apply_ufunc(_zhist, v, lat, kwargs={'weights': True},
                    output_core_dims=[['latitude']], output_dtypes=[np.float64], 
                    dask='parallelized',
                    input_core_dims=[dims, dims],
                    output_sizes={'latitude': len(latc)})
    S = S.assign_coords(latitude=latc).rename('S_SSUV')
    #
    W = xr.apply_ufunc(_zhist, v, lat, kwargs={'weights': False},
                    output_core_dims=[['latitude']], output_dtypes=[np.float64], 
                    dask='parallelized',
                    input_core_dims=[dims, dims],
                    output_sizes={'latitude': len(latc)})
    W = W.assign_coords(latitude=latc).rename('W_SSUV')    
    return xr.merge([S, W], compat='identical')
```

```python
def _get_E_total(x, ufunc=True, **kwargs):
    ax = -1 if ufunc else 0
    #
    dkwargs = {'window': 'boxcar', 'return_onesided': False, 
               'detrend': False, 'scaling': 'density'}
    dkwargs.update(kwargs)
    f, E = signal.welch(x, fs=24., axis=ax, **dkwargs)
    #
    if ufunc:
        return E
    else:
        return f, E

def get_E_total(v, f=None, **kwargs):
    v = v.chunk({'time': len(v.time)})
    if 'nperseg' in kwargs:
        Nb = kwargs['nperseg']
    else:
        Nb = 60*24
        kwargs['nperseg']= Nb
    if f is None:
        f, E = _get_E_total(v.values, ufunc=False, **kwargs)
        return f, E
    else:
        E = xr.apply_ufunc(_get_E_total, v,
                    dask='parallelized', output_dtypes=[np.float64],
                    input_core_dims=[['time']],
                    output_core_dims=[['freq_time']],
                    output_sizes={'freq_time': Nb}, kwargs=kwargs)
        return E.assign_coords(freq_time=f).sortby('freq_time')
```

```python
E_dir = '/work/ALT/swot/aval/syn/xy/momentum_balance/hourly_8640/'
S_dir = '/work/ALT/swot/aval/syn/xy/momentum_balance/hourly_spectra/'

# spatial strides for decimation
dij=4
overwrite=True
    
#for face in range(13):
for face in [10]:

    Efile = S_dir+'E_uvt_f%02d.zarr'%(face)

    if not os.path.isdir(Efile) or overwrite:

        # load data
        dsu = xr.concat([xr.open_zarr(E_dir+'du_dt_f%02d_t%04d.zarr'%(face, t)) for t in T], dim='time')
        dsv = xr.concat([xr.open_zarr(E_dir+'dv_dt_f%02d_t%04d.zarr'%(face, t)) for t in T], dim='time')
        
        dsu = dsu.chunk({'i': 47, 'j':24, 'time':8640})
        dsv = dsv.chunk({'i': 47, 'j':24, 'time':8640})
                       
        ds = xr.merge([dsu, dsv], compat='equals').assign_coords(**grd_rspec_dij.sel(face=face))
        ds['uv_t'] = ds['du_dt'].fillna(0.) + 1j * ds['dv_dt'].fillna(0.)  
        print(ds)
        print('dataset size: %.1f GB' %(ds.nbytes / 1e9))

        # spectral computation
        # need to run once to get frequency line
        f, E = get_E_total(ds['uv_t'].isel(i=2,j=2))
        #
        E = get_E_total(ds['uv_t'], f=f)         

        # store
        E = E.to_dataset()
        for c in E.coords:
            try:
                del E[c].encoding['chunks']
            except:
                print(c)        
        E = E.chunk({'i': 24*8, 'j':47*4})
        #print(E)
        %time E.to_zarr(Efile, mode='w')

        print('--- face %d done'%face)

    else:
        print('--- face %d allready computed'%face)
```

##  band-integral results (KE)
### band integral of spectra

```python
face_all = [10]


for face in face_all:
    
    ds = xr.open_zarr(S_dir+'E_uvt_f%02d.zarr'%(face))
    chunks={'i': 188, 'j':192, 'freq_time':1440}
    ds = ds.chunk(chunks)     
    E = ds.uv_t
    #E = E.assign_coords(freq_cps=E.freq_time / 86400.) 
    df=E.freq_time[1]-E.freq_time[0]

    # total KE
    E_all = (E*df).sum(dim='freq_time')
    E_all = E_all.chunk({'i': 1080, 'j': 1080}) 
    E_all = np.real(E_all).rename('E_all')
    # high-frequency KE (<0.5 day)
    E_high = (E.where((abs(E.freq_time)>0.5), other=0.)*df).sum(dim='freq_time')
    E_high = E_high.chunk({'i': 1080, 'j': 1080}) 
    E_high = np.real(E_high).rename('E_high')
    # semi-diurnal
    E_semi = (E.where((abs(E.freq_time)>1.9) & (abs(E.freq_time)<2.1), other=0.)*df).sum(dim='freq_time')
    E_semi = E_semi.chunk({'i': 1080, 'j': 1080}) 
    E_semi = np.real(E_semi).rename('E_semi')
    # diurnal
    E_diurnal = (E.where((abs(E.freq_time)>0.9) & (abs(E.freq_time)<1.1), other=0.)*df).sum(dim='freq_time')
    E_diurnal = E_diurnal.chunk({'i': 1080, 'j': 1080}) 
    E_diurnal = np.real(E_diurnal).rename('E_diurnal')
    # near-inertial
    f_llc = 2.*2.*np.pi/86400. * np.sin(np.deg2rad(E.YC))
    f_cpd_llc = f_llc*86400/2./np.pi
    E_f = (E.where((abs(E.freq_time+f_cpd_llc)<0.1), other=0.)*df).sum(dim='freq_time')
    E_f = np.real(E_f).rename('E_f')
    # low-frequency KE (>0.5 day)    
    E_low = E_all - E_high
    E_low = E_low.chunk({'i': 1080, 'j': 1080}) 
    E_low = np.real(E_low).rename('E_low')
    
    # merge together
    E_band = xr.merge([E_high, E_semi, E_diurnal, E_f, E_all, E_low])
    
    file_E = S_dir+'E_uvt_band_integral_f%02d.zarr'%face
    E_band.to_zarr(file_E, mode='w')  
    print('face %d done'%face )
```
