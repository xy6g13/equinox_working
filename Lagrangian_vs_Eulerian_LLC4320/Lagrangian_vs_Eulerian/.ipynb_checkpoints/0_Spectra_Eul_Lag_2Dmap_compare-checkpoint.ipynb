# compare Lagrangian and Eulerian spectra
### last update: 2023.08.08

```python
root_dir = '/home/datawork-lops-osi/equinox/mit4320/parcels/'
run_name = 'global_T365j_dt1j_dij50/'
p = pa.parcels_output(root_dir+run_name, parquets=['trajectory'])
```

```python
E = p.load_diagnostic('spectra_uv_T60_mean0_lon_lat.zarr').rename({'E':'E_Lagrangian'})
E_L = E.E_Lagrangian
E_L
```

```python
E_Eulerian = xr.open_zarr(os.path.join(ut.root_data_dir,'diags/SSUV_T60_lat_lon.zarr')).rename({'E_SSUV': 'E_Eulerian'})
E_E = E_Eulerian
E_E
```

```python
ds = xr.merge([E_L, E_E])
ds
```

---
## Maps

```python
freq_sel = 4

fig, ax = plt.subplots(figsize=(36,6))
ax = plt.subplot(131)
np.log10(ds.E_Lagrangian.sel(frequency=freq_sel)).plot(x='lon_bins',cmap='viridis', vmin=-5, vmax=-1)

ax = plt.subplot(132)
np.log10(ds.E_Eulerian.sel(frequency=freq_sel)).plot(x='lon_bins',cmap='viridis', vmin=-5, vmax=-1)

ax = plt.subplot(133)
np.log10(ds.E_Lagrangian.sel(frequency=freq_sel)/ds.E_Eulerian.sel(frequency=freq_sel)).plot(x='lon_bins',cmap=cm.balance, vmin=-1, vmax=1)
```
